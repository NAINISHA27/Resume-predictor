CLEANING DATA
1. URLs,
2. hashtage,
3. mentions,
4. special letters,
5. punctuations,

```python
import re

def cleanResume(txt):
    try:
        # Clean the text
        cleanTxt = re.sub(r'http\S+', ' ', txt)
        cleanTxt = re.sub(r'RT\s*', ' ', cleanTxt)
        cleanTxt = re.sub(r'#\S+', ' ', cleanTxt)
        cleanTxt = re.sub(r'@\S+', ' ', cleanTxt)
        cleanTxt = re.sub(r'[!"#$%&\'()*+,-./:;<=>?@[\\\]^_`{|}~]', ' ', cleanTxt)
        cleanTxt = re.sub(r'[^\x00-\x7F]+', ' ', cleanTxt)
        cleanTxt = re.sub(r'\s+', ' ', cleanTxt).strip()
        
        return cleanTxt
    except Exception as e:
        print(f"Error processing text: {txt}")
        print(e)
        return txt  # Return the original text if there's an error
```

```python
df['Resume']=df['Resume'].apply(lambda x: cleanResume(x))
```

Words into categorical Values 

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category']=le.transform(df['Category'])
```

```python
df.Category.unique()
```

Vectorization

tfidf = Term Frequency-Inverse Document Frequency.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(df['Resume'])
requiredTaxt = tfidf.transform(df['Resume'])
```

```python
df
```

```python
requiredTaxt 
```

Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test =train_test_split(requiredTaxt,df['Category'],test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

Now let's train the model and print the classification report:

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

clf= OneVsRestClassifier(KNeighborsClassifier())
clf.fit(x_train, y_train)
ypred = clf.predict(x_test)
print(accuracy_score(y_test,ypred))
```
